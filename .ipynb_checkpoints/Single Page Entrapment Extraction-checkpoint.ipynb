```python
import pandas as pd
from tabula import read_pdf
import tabula
import re
```

```python
pdf = '190923_1000_Transit.pdf'
page = 385
```

```python
df = tabula.read_pdf(pdf, pages=page)

# df_revision = df.copy()
df_null = df.isna()

df.replace(to_replace=[r"\\t|\\n|\\r", "\t|\n|\r"], value=[" "," "], regex=True, inplace=True)
```

## Find Units

```python
units = []
for i in range(len(df)):
    if re.match(string = str(df[df.columns[0]][i]), pattern = r'E[SL]\d{3}'):
        units.append(df[df.columns[0]][i])
units
```

## Find Subway Stations

```python
stations = []
for i in range(len(df)):
    if type(df[df.columns[1]][i]) is not float:
        if len(df[df.columns[1]][i].split(' ')[:-1]) > 0:
            stations.append(' '.join(df[df.columns[1]][i].split(' ')[:-1]))
stations
```

## Find Subway Lines

```python
lines = []
for i in range(len(df)):
#     if type(df[df.columns[1]][i]) is not float and re.match(string = str(df[df.columns[1]][i]).split(' ')[-1], pattern = r'[ABCDEFGJLMNRQWZ1-7]') and re.match(string = str(df[df.columns[1]][i]).split(' ')[-1], pattern = r'^((?!Location).)*$'):
    if type(df[df.columns[1]][i]) is not float and re.match(string = str(df[df.columns[1]][i]).split(' ')[-1], pattern = r'[ABCDEFGJLMNRQWZ1-7]') and re.match(string = str(df[df.columns[1]][i]).split(' ')[-1], pattern = r'^((?![A-Z][a-z]).)*$'):
        lines.append(' '.join(df[df.columns[1]][i].split(' ')[-1:]))
lines
```

## Find the Number of Entrapments

```python
entrapments = []
for i in range(len(df)):
    if type(df[df.columns[2]][i]) is not float and re.match(string = str(df[df.columns[2]][i]).split(' ')[0], pattern = r'[0-9]'):
        entrapments.append(' '.join(df[df.columns[2]][i].split(' ')[0]))
entrapments
```

## Find Entrapment Details

```python
entrapment_list_text = []
for i in range(len(df)):
    if re.match(string = str(df[df.columns[2]][i]), pattern = r'^((?!Entrapments|#  of Comments).)*$') and type(df[df.columns[2]][i]) is not float:
        if type(df[df.columns[3]][i]) is not float:
            if re.match(string = str(df[df.columns[2]][i]).split(' ')[0], pattern = r'[0-9]'):
                entrapment_list_text.append(df[df.columns[2]][i] + ' ' + df[df.columns[3]][i])
            if re.match(string = str(df[df.columns[2]][i]).split(' ')[0], pattern = r'[^0-9]'):
                entrapment_list_text.append(df[df.columns[2]][i] + ' ' + df[df.columns[3]][i])
        else:
            if re.match(string = str(df[df.columns[2]][i]).split(' ')[0], pattern = r'[0-9]'):
                entrapment_list_text.append(df[df.columns[2]][i])
            if re.match(string = str(df[df.columns[2]][i]).split(' ')[0], pattern = r'[^0-9]'):
                entrapment_list_text.append(df[df.columns[2]][i])
```

```python
entrapment_detail = []
entrapment_text = ''
for i in range(len(entrapment_list_text)):
    if re.match(string = str(entrapment_list_text[i]), pattern = r'^[0-9]\ '):
        if len(entrapment_text) == 0:
            entrapment_text += entrapment_list_text[i][2:]
        else:
            entrapment_detail.append(entrapment_text)
            entrapment_text = ''
            entrapment_text += entrapment_list_text[i][2:]
#         print('starts with number ' + str(len(entrapment_text)))
    if re.match(string = str(entrapment_list_text[i]).split(' ')[0], pattern = r'[^0-9]'):
        entrapment_text += ' ' + entrapment_list_text[i]
#         print("doesn't start with number " + str(len(entrapment_text)))
    if i == (len(entrapment_list_text) - 1):
        entrapment_detail.append(entrapment_text)
#         print("the end " + str(len(entrapment_text)))
entrapment_detail
```

```python
df_output = pd.DataFrame(
    {'unit': units,
     'station': stations,
     'line': lines,
     'entrapments' : entrapments,
     'comments' : entrapment_detail
    })
```
